```python
X = np.array(df.drop(['class'],1))
y = np.array(df['class'])
```

```python
def randomforestclassifier(X,y):
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2,random_state=0)
    forest = RandomForestClassifier(n_estimators=100, random_state=0)
    forest.fit(X_train, y_train)
    print("accuracy on training set: %f" % forest.score(X_train, y_train))
    print("accuracy on test set: %f" % forest.score(X_test, y_test))
    filename = 'final_model.pkl'
    pickle.dump(forest,open(filename,'wb'))
    
```

```python
randomforestclassifier(X,y)
```

```python
def svc(X,y):
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2,random_state=0)
    linear_svm = LinearSVC().fit(X, y)
    print("accuracy on training set: %f" % linear_svm.score(X_train, y_train))
    print("accuracy on test set: %f" % linear_svm.score(X_test, y_test))
    
```

```python
svc(X,y)
```

```python
def logisticregression(X,y):
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2,random_state=0)
    logistic_regression = LogisticRegression().fit(X_train, y_train)
    print("accuracy on training set: %f" % logistic_regression.score(X_train, y_train))
    print("accuracy on test set: %f" % logistic_regression.score(X_test, y_test))
```

```python
logisticregression(X,y)
```

```python
def decisiontree(X,y):
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2,random_state=0)
    tree = DecisionTreeClassifier(random_state=0)
    tree.fit(X_train,y_train)
    print("accuracy on training set: %f" % tree.score(X_train, y_train))
    print("accuracy on test set: %f" % tree.score(X_test, y_test))
```

```python
decisiontree(X,y)
```

```python
def predictor(dicty):
    all_list = []
    all_list = dicty.values()
    pred = np.array(list(all_list))
    pred = pred.reshape(1,-1)
    with open('final_model.pkl', 'rb') as fid:
        forest = pickle.load(fid)
    
    predi = forest.predict(pred)
    
    if predi[0] == 1:
        result = "Good"
    else:
        result= "bad"
    print("Credit risk:"result)
```
